```python
import pandas as pd
```

```python
train = pd.read_csv('SalaryData_Train.csv')
```

```python
test = pd.read_csv('SalaryData_Test.csv')
```

```python
train.columns
```

```python
train.info()
```

```python
string_columns = ["workclass","education","maritalstatus","occupation","relationship","race","sex","native","Salary"]
```

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
for i in string_columns:
    train[i] = le.fit_transform(train[i])
    test[i] = le.fit_transform(test[i])
```

```python
train
```

```python
test
```

```python
features = train.iloc[:,:13]
label = train.iloc[:,13]
```

```python
#import categorical naive bayes model
from sklearn.naive_bayes import CategoricalNB
model = CategoricalNB()
model.fit(features,label)
```

```python
#for test data
from sklearn.metrics import accuracy_score, confusion_matrix
features1 = test.iloc[:,:13]
label1 = test.iloc[:,13]

y_pred = model.predict(features1)
acc = accuracy_score(label1,y_pred)*100
print("accuracy score=",acc) #85.6%

```

```python
confusion_matrix(label1, y_pred)
```
